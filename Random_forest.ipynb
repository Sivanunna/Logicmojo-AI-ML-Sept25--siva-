```
y = df['Churn']
x = df.drop('Churn', axis=1)
```

```
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```
from sklearn.preprocessing import LabelEncoder
lb = LabelEncoder()

for col in x_train.columns:
  if(x_train[col].dtype == 'object'):
      x_train[col] = lb.fit_transform(x_train[col])
      x_test[col] = lb.fit_transform(x_test[col])
```

```
y_train.value_counts()
```

```
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier(random_state=1)
dt.fit(x_train, y_train)
```

```
from sklearn.metrics import classification_report
y_pred = dt.predict(x_test)
result = classification_report(y_test, y_pred)
print(result)
```

```
y_pred_train = dt.predict(x_train)
result = classification_report(y_train, y_pred_train)
print(result)
```

```
from sklearn.ensemble import RandomForestClassifier, BaggingClassifier
```

```
bag_class = BaggingClassifier(
    estimator=DecisionTreeClassifier(),
    n_estimators=50,
    random_state=7,
    oob_score=True
)

```

```
bag_model = bag_class.fit(x_train,y_train)
```

```
bag_model.oob_score_
```

```
y_pred_bag_train = bag_model.predict(x_train)
result_bag_train = classification_report(y_train, y_pred_bag_train)
print(result_bag_train)
```

```
y_pred_bag_test = bag_model.predict(x_test)
result_bag_test = classification_report(y_test, y_pred_bag_test)
print(result_bag_test)
```

```
from sklearn.ensemble import RandomForestClassifier
```

```
rf_classifier = RandomForestClassifier(
    n_estimators=25,  # Number of decision trees in the forest
    max_features=6,  # Maximum number of features considered for splitting
    criterion='gini',  # The function to measure the quality of a split (Gini impurity)
    max_depth=10,  # Maximum depth of the decision trees
    min_impurity_decrease=0.05,  # Minimum impurity decrease required for a split
    bootstrap=True,  # Whether bootstrap samples are used when building trees
    oob_score=True,  # Whether to use out-of-bag samples to estimate generalization accuracy
    class_weight='balanced',  # Weights associated with classes to handle class imbalance
    random_state=7  # Controls the random seed for reproducibility
)
```

```
rf_model = rf_classifier.fit(x_train, y_train)
```

```
y_pred_bag_train1 = rf_model.predict(x_train)
result_bag_train1 = classification_report(y_train, y_pred_bag_train1)
print(result_bag_train1)
```

```
y_pred_bag_train1 = rf_model.predict(x_test)
result_bag_train1 = classification_report(y_test, y_pred_bag_train1)
print(result_bag_train1)
```

```
from sklearn.model_selection import GridSearchCV
```

```
params = {
    "n_estimators": [25, 50, 5, 100],  # Number of decision trees in the forest
    "criterion": ["entropy", "gini"],  # The function to measure the quality of a split
    "max_depth": [5, 10, 12, 15],  # Maximum depth of the decision trees
    "max_features": [5, 6, 7, 8]  # Maximum number of features considered for splitting
}
```

```
grid_model = GridSearchCV(
    estimator=RandomForestClassifier(),  # The base estimator to be tuned, in this case, RandomForestClassifier
    param_grid=params,  # The parameter grid containing hyperparameter values to be searched
    scoring='f1',  # The scoring metric to optimize for during hyperparameter tuning, in this case, F1-score
    cv=10,  # Number of folds for cross-validation
    verbose=3,  # Controls the verbosity of the grid search process
    n_jobs=-1  # Number of jobs to run in parallel (-1 means using all available cores)
)
```

```
grid_model.fit(x_train, y_train)
```

```
# Access the best estimator found by GridSearchCV
best_rf_estimator = grid_model.best_estimator_
best_rf_estimator
```

```
print("Best Estimator: \n{}\n".format(grid_model.best_estimator_))

# Print the best parameters found by GridSearchCV
print("Best Parameter: \n{}\n".format(grid_model.best_params_))
```
